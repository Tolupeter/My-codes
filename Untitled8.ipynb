```python
import scanpy as sc
import scvi
import seaborn as sns
import matplotlib.pyplot as plt

# Set working directory (adjust this path to your data directory)
os.chdir("P:/Tolulope/Manuscript/Yuan Analysis")

# Load each sample's data one by one
adata1 = sc.read_h5ad("YSham_GFP_adata.h5ad")
adata2 = sc.read_h5ad("YSham_noGFP_adata.h5ad")
adata3 = sc.read_h5ad("YOV_GFP_adata.h5ad")
adata4 = sc.read_h5ad("YOV_noGFP_adata.h5ad")
adata5 = sc.read_h5ad("ASham_GFP_adata.h5ad")
adata6 = sc.read_h5ad("ASham_noGFP_adata.h5ad")
adata7 = sc.read_h5ad("AOV_GFP_adata.h5ad")
adata8 = sc.read_h5ad("AOV_noGFP_adata.h5ad")

print("All datasets loaded successfully!")
def process_doublets_with_SOLO(adata, sample_name):
    print(f"Processing {sample_name} - Initial cell count: {adata.shape[0]}")
    
    # Filter genes expressed in at least 3 cells
    sc.pp.filter_genes(adata, min_cells=3)
    print(f"Cell count after gene filtering: {adata.shape[0]}")
    
    # Identify highly variable genes
    sc.pp.highly_variable_genes(adata, n_top_genes=3000, subset=True, flavor='seurat_v3')
    print(f"Cell count after selecting highly variable genes: {adata.shape[0]}")
    
    # Train scVI model
    scvi.model.SCVI.setup_anndata(adata)
    vae = scvi.model.SCVI(adata)
    vae.train()
    
    # Run SOLO for doublet detection
    solo = scvi.external.SOLO.from_scvi_model(vae)
    solo.train()
    
    # Predict doublets
    df = solo.predict()
    df['prediction'] = solo.predict(soft=False)
    df['dif'] = df.doublet - df.singlet
    
    # Plot confidence difference for doublets
    sns.displot(df[df['prediction'] == 'doublet'], x='dif')
    plt.xlabel('Confidence Difference (doublet - singlet)')
    plt.title(f'{sample_name} - Doublet Confidence Difference')
    plt.show()
    
    # Set threshold for doublet filtering
    threshold = float(input(f"Enter confidence difference threshold for {sample_name} (e.g., 0.1): "))
    
    # Adjust indices for matching
    adata.obs['adjusted_index'] = adata.obs.index.str.replace(f"-1-{sample_name}", "")
    df.index = df.index.str.replace(f"-1-{sample_name}", "")
    
    # Mark doublets based on threshold
    doublet_indices = df[df['dif'] > threshold].index
    adata.obs['doublet'] = adata.obs['adjusted_index'].isin(doublet_indices)
    
    # Filter out doublets
    adata_filtered = adata[~adata.obs['doublet']].copy()
    print(f"{sample_name} - Remaining cells after doublet removal: {adata_filtered.shape[0]}")
    
    return adata_filtered

# Process each dataset
samples = {
    "YSham_GFP": adata1,
    "YSham_noGFP": adata2,
    "YOV_GFP": adata3,
    "YOV_noGFP": adata4,
    "ASham_GFP": adata5,
    "ASham_noGFP": adata6,
    "AOV_GFP": adata7,
    "AOV_noGFP": adata8
}

filtered_data = {sample: process_doublets_with_SOLO(adata, sample) for sample, adata in samples.items()}

# Save filtered datasets
for sample, adata in filtered_data.items():
    adata.write(f"{sample}_adata_filtered_with_SOLO.h5ad")

print("All filtered datasets saved successfully!")
```

```python
import os
import scanpy as sc

# Set working directory (adjust this path to your data directory)
os.chdir("P:/Tolulope/Manuscript/Yuan Analysis")

# Define sample file names
sample_files = [
    "YSham_GFP_adata_filtered_with_SOLO.h5ad",
    "YSham_noGFP_adata_filtered_with_SOLO.h5ad",
    "YOV_GFP_adata_filtered_with_SOLO.h5ad",
    "YOV_noGFP_adata_filtered_with_SOLO.h5ad",
    "ASham_GFP_adata_filtered_with_SOLO.h5ad",
    "ASham_noGFP_adata_filtered_with_SOLO.h5ad",
    "AOV_GFP_adata_filtered_with_SOLO.h5ad",
    "AOV_noGFP_adata_filtered_with_SOLO.h5ad"
]

# Load datasets into a list
adata_list = [sc.read_h5ad(file) for file in sample_files]

print("All datasets loaded successfully!")


def preprocess_adata(adata):
    """Preprocesses AnnData object by filtering low-quality cells and annotating genes."""
    sc.pp.filter_cells(adata, min_genes=200)
    
    # Annotate hemoglobin (Hb) and mitochondrial (mt) genes
    adata.var['hb'] = adata.var_names.str.contains(r'^Hb(?!P)', regex=True)  # Exclude HbP
    adata.var['mt'] = adata.var_names.str.startswith('mt-')

    # Compute QC metrics
    sc.pp.calculate_qc_metrics(adata, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)

    # Filter cells based on quality thresholds
    adata = adata[adata.obs.n_genes_by_counts < 3000, :]
    adata = adata[adata.obs.pct_counts_mt < 5, :]

    return adata


# Apply preprocessing to each dataset
adata_list = [preprocess_adata(adata) for adata in adata_list]

# Concatenate all processed datasets
adata_combined = sc.concat(adata_list)

# Save the combined dataset
adata_combined.write_h5ad("combined.h5ad")

print("Preprocessing complete. Combined dataset saved as 'combined.h5ad'!")
```
