```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os
import re
import os
# Set working directory
os.chdir("P:/Tolulope/Manuscript/Yuan Analysis")
# Function to calculate M2 score
def calculate_m2_score(expr_data, m2_genes, transpose=False):
    if transpose:
        expr_data_t = expr_data.copy()
    else:
        expr_data_t = expr_data.T.copy()
    
    gene_names = expr_data_t.index.tolist()
    
    available_m2_genes = []
    for gene in m2_genes:
        if gene in gene_names:
            available_m2_genes.append(gene)
        else:
            gene_lower = gene.lower()
            for g in gene_names:
                if isinstance(g, str) and gene_lower == g.lower():
                    available_m2_genes.append(g)
                    break
    
    if not available_m2_genes:
        print("No M2 marker genes found in the dataset!")
        return pd.Series(np.zeros(expr_data_t.shape[1]), index=expr_data_t.columns)
    
    m2_expr = expr_data_t.loc[available_m2_genes]
    m2_scores = m2_expr.mean(axis=0)
    return m2_scores

# ECDF function
def ECDF_standard(data, ax=None, **kwargs):
    x = np.sort(data)
    y = np.arange(1, len(x)+1) / len(x)
    if ax is None:
        plt.plot(x, y, **kwargs)
    else:
        ax.plot(x, y, **kwargs)
    return x, y

# M2 polarization analysis function
def analyze_m2_polarization():
    print("Loading datasets...")
    try:
        pseudotime_df = pd.read_csv("pseudotime_metadata_for_python.csv", index_col=0)
        print(f"Pseudotime metadata loaded. Shape: {pseudotime_df.shape}")
        normalized_expr = pd.read_csv("normalized_expr_matrix_for_python.csv", index_col=0)
        print(f"Normalized expression matrix loaded. Shape: {normalized_expr.shape}")
        print("\nSample of expression matrix:")
        print(f"First few row names: {normalized_expr.index[:5].tolist()}")
        print(f"First few column names: {normalized_expr.columns[:5].tolist()}")
        
        first_col = normalized_expr.columns[0]
        has_cell_barcodes = any(re.search(r'\w+\.\d+\.\w+', str(col)) for col in normalized_expr.columns[:10])
        
        if has_cell_barcodes:
            print("\nDetected that expression matrix has genes as rows and cells as columns")
            genes_are_rows = True
        else:
            print("\nDetected that expression matrix has cells as rows and genes as columns")
            genes_are_rows = False
        
        try:
            cell_metadata = pd.read_csv("cell_metadata_for_python.csv", index_col=0)
            print(f"Cell metadata loaded. Shape: {cell_metadata.shape}")
        except FileNotFoundError:
            cell_metadata = None
            print("Cell metadata file not found, continuing without it.")
        
        try:
            gene_anno = pd.read_csv("gene_annotation_for_python.csv", index_col=0)
            print(f"Gene annotation loaded. Number of genes: {gene_anno.shape[0]}")
            if 'gene_symbol' in gene_anno.columns:
                print(f"Sample gene symbols: {gene_anno['gene_symbol'].head(10).tolist()}")
        except FileNotFoundError:
            gene_anno = None
            print("Gene annotation file not found, continuing without it.")
        
        cluster_pseudotime = pd.read_csv("cluster_pseudotime_for_python.csv")
        print(f"Cluster pseudotime data loaded. Shape: {cluster_pseudotime.shape}")
        print(f"Sample columns: {cluster_pseudotime.columns.tolist()}")
        
    except FileNotFoundError as e:
        print(f"Error loading data files: {e}")
        return
    
    m2_markers = [
        'Arg1', 'Mrc1', 'CD206', 'Retnla', 'Fizz1', 'Chi3l3', 'Ym1', 'Il10', 
        'Tgfb1', 'Cd163', 'Mgl1', 'Mgl2', 'Ccl17', 'Ccl22', 'Tgm2', 'Fn1',
        'Msr1', 'Marco', 'Folr2', 'Clec7a', 'Igf1', 'Pdgfc'
    ]
    
    if genes_are_rows:
        available_genes = normalized_expr.index.tolist()
    else:
        available_genes = normalized_expr.columns.tolist()
    
    available_m2_markers = []
    for marker in m2_markers:
        exact_matches = [g for g in available_genes if marker == g]
        if exact_matches:
            available_m2_markers.extend(exact_matches)
            continue
            
        marker_lower = marker.lower()
        for g in available_genes:
            if isinstance(g, str) and marker_lower == g.lower():
                available_m2_markers.append(g)
                break
    
    if not available_m2_markers and gene_anno is not None:
        if 'gene_symbol' in gene_anno.columns:
            for index, row in gene_anno.iterrows():
                symbol = row['gene_symbol']
                if isinstance(symbol, str):
                    if any(marker.lower() in symbol.lower() for marker in m2_markers):
                        if index in available_genes:
                            available_m2_markers.append(index)
                        elif genes_are_rows and symbol in normalized_expr.index:
                            available_m2_markers.append(symbol)
                        elif not genes_are_rows and symbol in normalized_expr.columns:
                            available_m2_markers.append(symbol)
    
    if not available_m2_markers:
        print("No matches found. Please provide custom M2 markers for your dataset.")
        
    print(f"\nFound {len(available_m2_markers)} M2 markers in the dataset: {available_m2_markers}")
    
    if not available_m2_markers:
        print("Warning: No M2 markers found in the dataset!")
        return
    
    m2_scores = calculate_m2_score(normalized_expr, available_m2_markers, transpose=genes_are_rows)
    m2_scores_df = pd.DataFrame({'M2_Score': m2_scores})
    print(f"M2 scores calculated for {len(m2_scores)} cells")
    
    print("\nMerging M2 scores with pseudotime data...")
    
    if 'Sample' in cluster_pseudotime.columns:
        cell_ids_in_pseudotime = set(cluster_pseudotime['Sample'].values)
        cell_ids_in_m2scores = set(m2_scores_df.index)
        overlap = cell_ids_in_pseudotime.intersection(cell_ids_in_m2scores)
        print(f"Found {len(overlap)} cells that match between M2 scores and pseudotime data")
        
        if len(overlap) == 0:
            if genes_are_rows:
                pattern = r'([^\.]+)\.([^\.]+)\.([^\.]+)'
                cell_id_map = {}
                
                for col in normalized_expr.columns:
                    match = re.search(pattern, str(col))
                    if match:
                        barcode, batch, sample = match.groups()
                        cell_id_map[col] = barcode
                
                mapped_m2_scores = pd.Series(index=[cell_id_map.get(col, col) for col in m2_scores.index], 
                                           data=m2_scores.values)
                m2_scores_df = pd.DataFrame({'M2_Score': mapped_m2_scores})
                
                cell_ids_in_m2scores = set(m2_scores_df.index)
                overlap = cell_ids_in_pseudotime.intersection(cell_ids_in_m2scores)
                print(f"After ID mapping: {len(overlap)} cells match between datasets")
        
        if len(overlap) > 0:
            analysis_df = cluster_pseudotime.copy()
            analysis_df = analysis_df.set_index('Sample')
            common_cells = list(overlap)
            analysis_df = analysis_df.loc[common_cells]
            m2_scores_subset = m2_scores_df.loc[common_cells]
            analysis_df['M2_Score'] = m2_scores_subset['M2_Score']
            analysis_df = analysis_df.reset_index()
        else:
            print("Warning: Could not match cell IDs between datasets!")
            return
    else:
        print("Error: Pseudotime data doesn't have a 'Sample' column!")
        return
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    colors = ["#17becf", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
    sample_groups = np.unique(analysis_df['Sample'].apply(lambda x: x.split('.')[-1] if '.' in str(x) else x))
    analysis_df['SampleGroup'] = analysis_df['Sample'].apply(lambda x: x.split('.')[-1] if '.' in str(x) else x)
    
    results = {}
    for i, group in enumerate(sample_groups):
        group_data = analysis_df[analysis_df['SampleGroup'] == group]
        
        if len(group_data) == 0:
            print(f"Warning: No data for group {group}")
            continue
        
        axes[0, 0].scatter(
            group_data['Pseudotime'], 
            group_data['M2_Score'],
            alpha=0.6, 
        )
        axes[0, 0].set_xlabel('Pseudotime')
        axes[0, 0].set_ylabel('M2 Polarization Score')
        axes[0, 0].set_title(f'M2 Score vs Pseudotime: {group}')

        sorted_data = group_data.sort_values('Pseudotime')
        axes[1, 0].plot(
            sorted_data['Pseudotime'], 
            sorted_data['M2_Score'], 
            label=f'{group} Trend', 
            color=colors[i % len(colors)]
        )
        axes[1, 0].set_xlabel('Pseudotime')
        axes[1, 0].set_ylabel('M2 Polarization Score')
        axes[1, 0].set_title(f'M2 Score Trend: {group}')
    
        print(f"\nAnalysis for Group {group}:")
        print("Descriptive statistics:")
        print(group_data['M2_Score'].describe())
        
        # Perform linear regression on the data
        slope, intercept, r_value, p_value, std_err = stats.linregress(group_data['Pseudotime'], group_data['M2_Score'])
        print(f"Linear regression results: slope={slope:.3f}, p-value={p_value:.4e}")
        
        results[group] = {
            'slope': slope,
            'intercept': intercept,
            'r_value': r_value,
            'p_value': p_value,
            'std_err': std_err,
        }
    
    axes[0, 0].legend(title="Sample Groups")
    plt.tight_layout()
    plt.show()

    return results
```
