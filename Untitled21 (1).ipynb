```python
import scanpy as sc
import seaborn as sns
import numpy as np
import pandas as pd
import random
import os
from matplotlib.pyplot import rc_context
sc.set_figure_params(dpi=100)

import warnings
warnings.simplefilter("ignore", FutureWarning)
warnings.simplefilter("ignore", UserWarning)
warnings.simplefilter("ignore", RuntimeWarning)
# Set working directory
os.chdir("P:/Tolulope/Manuscript/Yuan Analysis")

```

```python
adata = sc.read_h5ad('integratedssssss.h5ad')
adata
```

```python
adata.raw.X
```

```python
cell_subset = adata[adata.obs['cell type'] == 'Lipid-Laden Macrophages']
cell_subset
```

```python
#example WITH pseudo replicates
pbs = []
for sample in cell_subset.obs.Sample.unique():
    samp_cell_subset = cell_subset[cell_subset.obs['Sample'] == sample]
    
    samp_cell_subset.X = samp_cell_subset.layers['counts'] #make sure to use raw data
    
    
    
    indices = list(samp_cell_subset.obs_names)
    random.shuffle(indices)
    indices = np.array_split(np.array(indices), 4) #change number here for number of replicates deisred
    
    for i, pseudo_rep in enumerate(indices):
    
        rep_adata = sc.AnnData(X = samp_cell_subset[indices[i]].X.sum(axis = 0),
                               var = samp_cell_subset[indices[i]].var[[]])

        rep_adata.obs_names = [sample + '_' + str(i)]
        rep_adata.obs['condition'] = samp_cell_subset.obs['condition'].iloc[0]
        rep_adata.obs['replicate'] = i

        pbs.append(rep_adata)
```

```python
pb = sc.concat(pbs)
```

```python
pb.obs
```

```python
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
```

```python
counts = pd.DataFrame(pb.X, columns = pb.var_names) #need to do this to pass var names
```

```python

# Create DeseqDataSet object with pseudo-replicates
dds = DeseqDataSet(
    counts=counts,  # Make sure 'counts' is correctly defined
    metadata=pb.obs,
    design_factors=['condition', 'replicate']  # Adjust to 'pseudo_replicate_group' if needed
)
```

```python
dds
```

```python
sc.pp.filter_genes(dds, min_cells = 1)
```

```python
dds
```

```python
dds.deseq2()
```

```python
sc.set_figure_params(figsize=(10, 10))
img_tag = "_color"  # Specify your image tag here
sc.tl.pca(dds)
sc.pl.pca(dds, color = 'condition', size = 200, save=f"pca_M2I_{img_tag}.png")
```

```python
stat_res = DeseqStats(dds, contrast=('condition', 'YOV-GFP', 'AOV-GFP'))
    
stat_res.summary()
```

```python
import pandas as pd

df = pd.read_csv("YOV-GFP_vs_AOV-GFP_LIP.csv")

logfc_threshold = 0.5
padj_threshold = 0.05

up = df[(df['log2FoldChange'] > logfc_threshold) & (df['padj'] < padj_threshold)]
down = df[(df['log2FoldChange'] < -logfc_threshold) & (df['padj'] < padj_threshold)]

print(f"Upregulated (log2FC > 0.5): {len(up)}")
print(f"Downregulated (log2FC < -0.5): {len(down)}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Filter for significant DEGs
sig_de = de[de['padj'] < 0.05].copy()
sig_de['log2FoldChange'] = sig_de['log2FoldChange'].astype(float)

# Assign angle based on up or downregulation
np.random.seed(42)
sig_de['angle'] = np.where(
    sig_de['log2FoldChange'] >= 0,
    np.random.uniform(-np.pi / 3, np.pi / 3, size=sig_de.shape[0]),   # right half
    np.random.uniform(2 * np.pi / 3, 4 * np.pi / 3, size=sig_de.shape[0])  # left half
)

# Compute radius with a bit of noise
sig_de['radius'] = sig_de['log2FoldChange'] + np.random.normal(0, 0.05, size=sig_de.shape[0])

# Top genes for labeling
top_up = sig_de[sig_de['log2FoldChange'] > 0].nlargest(10, 'log2FoldChange')
top_down = sig_de[sig_de['log2FoldChange'] < 0].nsmallest(8, 'log2FoldChange')

# Check if top_up or top_down is empty
if top_up.empty:
    print("Warning: No upregulated genes found for annotation.")
if top_down.empty:
    print("Warning: No downregulated genes found for annotation.")

# List of additional genes to label
additional_genes = ['Col1a1', 'Col1a2']
additional_genes_df = sig_de[sig_de.index.isin(additional_genes)]

# Check if additional genes exist in the dataset
missing_genes = [gene for gene in additional_genes if gene not in sig_de.index]
if missing_genes:
    print(f"Warning: These additional genes are missing: {missing_genes}")

# Combine the top 1 upregulated, 8 downregulated, and additional genes
top_genes = pd.concat([top_up, top_down, additional_genes_df])

# Check if top_genes is empty
if top_genes.empty:
    print("Warning: No genes to annotate after combining top upregulated, downregulated, and additional genes.")

# Plotting
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)

# Radius formatting
rmin = np.floor(sig_de['radius'].min())
rmax = np.ceil(sig_de['radius'].max())
ax.set_yticks(np.arange(rmin, rmax + 1, 1))
ax.set_ylim(rmin - 0.5, rmax + 0.5)
ax.set_yticklabels(np.arange(int(rmin), int(rmax) + 1), fontsize=11)
ax.set_xticks([])

# Scatter plot with custom colors
sig_de['color'] = np.where(sig_de['log2FoldChange'] >= 0, "#bb0c00", "#4169E1")  # Custom color scale
ax.scatter(sig_de['angle'], sig_de['radius'], 
           c=sig_de['color'], s=30, edgecolor='black', alpha=0.85)

# Annotate top genes with leader lines (only if top_genes is not empty)
if not top_genes.empty:
    for _, row in top_genes.iterrows():
        theta = row['angle']
        r = row['radius']
        label_angle = theta

        # Adjust label radius for downregulated genes to avoid overlap
        if row['log2FoldChange'] < 0:
            label_radius = r + 1.2 * np.sign(r)  # More offset for downregulated genes
            
            # Add check to avoid extreme label positions
            label_radius = min(max(label_radius, rmin - 0.5), rmax + 0.5)
        else:
            label_radius = r + 0.7 * np.sign(r)

        # Define arrow length for downregulated genes
        arrow_length = 0.9 if row['log2FoldChange'] >= 0 else 1.5  # Longer arrows for downregulated genes

        ha = 'left' if theta < np.pi else 'right'
        ax.annotate(row.name,
                    xy=(theta, r), 
                    xytext=(label_angle, label_radius),
                    textcoords='data',
                    fontsize=12,  # Increased font size
                    ha=ha,
                    va='center',
                    rotation=30 if row['log2FoldChange'] < 0 else 0,  # Rotate text for downregulated genes
                    color='black',  # Label color for better contrast
                    fontweight='bold',  # Bold labels for better visibility
                    bbox=dict(facecolor='white', edgecolor='none', alpha=0.7),  # Add background for better contrast
                    arrowprops=dict(arrowstyle="-", lw=1.2, color='gray', 
                                    connectionstyle=f"arc3,rad={arrow_length}"))

# Midline
ax.plot([np.pi / 2, 3 * np.pi / 2], [rmin - 0.5, rmax + 0.5],
        color='gray', linestyle='--', linewidth=1)

# Labels for circle halves
ax.text(0, rmax + 1, "AOV-GFP", fontsize=16, weight='bold',
        ha='center', color="#bb0c00")
ax.text(np.pi, rmax + 1, "AOV-noGFP", fontsize=16, weight='bold',
        ha='center', color="#4169E1")

# Title with more space
ax.set_title("DEGs in M2-like Macrophages\n(AOV-GFP vs AOV-noGFP)", 
             fontsize=18, pad=40)  # Increased padding for more space

# Adjust layout
plt.tight_layout()


# Save the figure
plt.savefig('m2AOV.png', dpi=300)  # Save the figure with higher resolution
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Filter for significant DEGs
sig_de = de[de['padj'] < 0.05].copy()
sig_de['log2FoldChange'] = sig_de['log2FoldChange'].astype(float)

# Assign angle based on up or downregulation
np.random.seed(42)
sig_de['angle'] = np.where(
    sig_de['log2FoldChange'] >= 0,
    np.random.uniform(-np.pi / 3, np.pi / 3, size=sig_de.shape[0]),   # right half
    np.random.uniform(2 * np.pi / 3, 4 * np.pi / 3, size=sig_de.shape[0])  # left half
)

# Compute radius with a bit of noise
sig_de['radius'] = sig_de['log2FoldChange'] + np.random.normal(0, 0.05, size=sig_de.shape[0])

# Top genes for labeling
top_up = sig_de[sig_de['log2FoldChange'] > 0].nlargest(10, 'log2FoldChange')
top_down = sig_de[sig_de['log2FoldChange'] < 0].nsmallest(8, 'log2FoldChange')

# Additional genes to label
additional_genes = ['Rpl3', 'Rps2', 'Ybx1', 'Rpl15', 'Rpl6', 'Rpl14', 'Rpl9', 
                    'Atp5f1c', 'Atp5pb', 'Atp5f1a', 'Ndufb5']
additional_genes_df = sig_de[sig_de.index.isin(additional_genes)]

# Combine the top 10 with additional genes
top_genes = pd.concat([top_up, top_down, additional_genes_df])

# Plotting
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)

# Radius formatting
rmin = np.floor(sig_de['radius'].min())
rmax = np.ceil(sig_de['radius'].max())
ax.set_yticks(np.arange(rmin, rmax + 1, 1))
ax.set_ylim(rmin - 0.5, rmax + 0.5)
ax.set_yticklabels(np.arange(int(rmin), int(rmax) + 1), fontsize=11)
ax.set_xticks([])

# Scatter plot with custom colors
sig_de['color'] = np.where(sig_de['log2FoldChange'] >= 0, "#bb0c00", "#4169E1")  # Custom color scale
ax.scatter(sig_de['angle'], sig_de['radius'], 
           c=sig_de['color'], s=30, edgecolor='black', alpha=0.85)

# Annotate top genes with leader lines
for _, row in top_genes.iterrows():
    theta = row['angle']
    r = row['radius']
    label_angle = theta
    
    # Adjust label radius for downregulated genes to avoid overlap
    if row['log2FoldChange'] < 0:
        label_radius = r + 1.2 * np.sign(r)  # More offset for downregulated genes
    else:
        label_radius = r + 0.7 * np.sign(r)
    
    # Define arrow length for downregulated genes
    arrow_length = 0.9 if row['log2FoldChange'] >= 0 else 1.5  # Longer arrows for downregulated genes
    
    ha = 'left' if theta < np.pi else 'right'
    ax.annotate(row.name,
                xy=(theta, r), 
                xytext=(label_angle, label_radius),
                textcoords='data',
                fontsize=12,  # Increased font size
                ha=ha,
                va='center',
                rotation=30 if row['log2FoldChange'] < 0 else 0,  # Rotate text for downregulated genes
                color='black',  # Label color for better contrast
                fontweight='bold',  # Bold labels for better visibility
                bbox=dict(facecolor='white', edgecolor='none', alpha=0.7),  # Add background for better contrast
                arrowprops=dict(arrowstyle="-", lw=1.2, color='gray', 
                                connectionstyle=f"arc3,rad={arrow_length}"))

# Midline
ax.plot([np.pi / 2, 3 * np.pi / 2], [rmin - 0.5, rmax + 0.5],
        color='gray', linestyle='--', linewidth=1)

# Labels for circle halves
ax.text(0, rmax + 1, "ASham-GFP", fontsize=16, weight='bold',
        ha='center', color="#bb0c00")
ax.text(np.pi, rmax + 1, "ASham-noGFP", fontsize=16, weight='bold',
        ha='center', color="#4169E1")

# Title with more space
ax.set_title("DEGs in M2-like Macrophages\n(ASham-GFP vs ASham-noGFP)", 
             fontsize=18, pad=40)  # Increased padding for more space

# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig('dendriticasham.png', dpi=300)  # Save the figure with higher resolution
plt.show()
```
