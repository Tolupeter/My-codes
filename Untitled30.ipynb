```python
import scanpy as sc
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.sparse
import random
import os

# Set working directory
os.chdir("P:/Tolulope/Manuscript/Yuan Analysis")
adata = sc.read_h5ad('integratedssssss.h5ad')
adata
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Remap condition names
def map_condition(x):
    condition_map = {
        'YSham_GFP': 'YSham-GFP',
        'YSham_noGFP': 'YSham-noGFP',
        'YOV_GFP': 'YOV-GFP',
        'YOV_noGFP': 'YOV-noGFP',
        'ASham_GFP': 'ASham-GFP',
        'ASham_noGFP': 'ASham-noGFP',
        'AOV_GFP': 'AOV-GFP',
        'AOV_noGFP': 'AOV-noGFP'
    }
    return condition_map.get(x, x)

# Map conditions
adata.obs['condition'] = adata.obs['batch'].map(map_condition)

# Select conditions of interest
selected_conditions = ['YSham-GFP', 'YSham-noGFP']

# Count cells per cell type and condition
cell_type_counts = (
    adata.obs[adata.obs['condition'].isin(selected_conditions)]
    .groupby(['cell type', 'condition'])
    .size()
    .reset_index(name='count')
)

# Pivot to wide format and normalize to frequency (0–1)
pivot_data = cell_type_counts.pivot(index='cell type', columns='condition', values='count').fillna(0)
pivot_data = pivot_data.div(pivot_data.sum(axis=1), axis=0)  # normalize to frequency
pivot_data = pivot_data.sort_index()

# Set up plot
fig, ax = plt.subplots(figsize=(8, 6))
cell_types = pivot_data.index
y_positions = range(len(cell_types))

# Define colors
colors = {
    'YSham-GFP': 'skyblue',
    'YSham-noGFP': 'steelblue'
}

# Plot YSham-GFP bars
ax.barh(
    y=y_positions,
    width=pivot_data['YSham-GFP'],
    color=colors['YSham-GFP'],
    edgecolor='black',
    label='YSham-GFP'
)

# Plot YSham-noGFP bars stacked on top
ax.barh(
    y=y_positions,
    width=pivot_data['YSham-noGFP'],
    left=pivot_data['YSham-GFP'],
    color=colors['YSham-noGFP'],
    hatch='//',
    edgecolor='black',
    label='YSham-noGFP'
)

# Formatting
ax.set_yticks(y_positions)
ax.set_yticklabels(cell_types, fontsize=10)
ax.set_xlabel('Prop', fontsize=12)
ax.set_xlim(0, 1)
#ax.set_title('Stacked Frequency of Cell Types: YSham-GFP vs ASham-GFP', fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=10)

# Legend outside the plot
ax.legend(title='Condition', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title_fontsize=11)

# Tidy layout
plt.tight_layout(rect=[0, 0, 0.8, 1])  # make room for legend
# Save figure
plt.savefig("cell_type_frequencies_YSham_vs_YShamno.png", dpi=300, bbox_inches='tight')
plt.show()
```
