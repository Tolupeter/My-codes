```python
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import os

# Set working directory
os.chdir("P:/Tolulope/Manuscript/Yuan Analysis")

# Load processed and integrated data
adata = sc.read_h5ad("integratedssssss.h5ad")

# Check data types and convert if necessary
print(adata.obs.dtypes)  # Check the types of columns in 'obs'
adata.obs['Sample'] = adata.obs['Sample'].astype('category')
adata.obs['cell type'] = adata.obs['cell type'].astype('category')

# Subset to M2-like macrophages from ASham condition (with and without exosome)
adata = adata[adata.obs["cell type"] == "M2-like macrophages", :]
adata = adata[adata.obs["Sample"].isin(["ASham_GFP", "ASham_noGFP"])].copy()

# Optional: check counts
print(adata.obs["Sample"].value_counts())

# 1. Subsetting GFP+ and GFP- cells
gfp_positive = adata[adata.obs['Sample'] == 'ASham_GFP', :]
gfp_negative = adata[adata.obs['Sample'] == 'ASham_noGFP', :]


```

```python
import numpy as np
import pandas as pd

# Perform Differential Gene Expression (DEG) analysis between GFP+ and GFP- cells
sc.tl.rank_genes_groups(adata, groupby='Sample', method='t-test', n_genes=25)  # Adjust n_genes as needed

# Get the results in a dataframe
deg_results = adata.uns['rank_genes_groups']['names']
deg_pvals = adata.uns['rank_genes_groups']['pvals']

# Print out the raw results for debugging
print("Raw DEG Results (Gene Names):")
print(deg_results[:25])  # View the first 25 genes

print("Raw DEG P-values:")
print(deg_pvals[:25])  # View the first 25 p-values

# Flatten the gene names (extract the first gene from each tuple)
flattened_genes = [gene[0] for gene in deg_results[:25]]  # Extract first gene of each tuple

# Flatten the p-values (select the minimum p-value for each gene pair)
flattened_pvals = [min(pvals) for pvals in deg_pvals[:25]]  # Use the smaller p-value from the tuple

# Convert to pandas DataFrame for easier handling
deg_results_df = pd.DataFrame({
    'Gene': flattened_genes,
    'p-value': flattened_pvals
})

# Drop any rows with NaN p-values
deg_results_df.dropna(subset=['p-value'], inplace=True)

# Now the DataFrame should be clean
print("Cleaned DEG Results:")
print(deg_results_df.head(10))
```

```python
# 4. Correlate gene expression with GFP status for selected genes
genes_of_interest = ['C1qa', 'C1qb', 'C1qc']  # Replace with your gene names

# Extract expression data for the genes of interest from adata.X
gfp_pos_expr = adata[adata.obs['Sample'] == 'ASham_GFP', genes_of_interest].X.toarray()
gfp_neg_expr = adata[adata.obs['Sample'] == 'ASham_noGFP', genes_of_interest].X.toarray()

# Correlate expression levels with GFP status
correlations = {}
for gene in genes_of_interest:
    # Expression levels for each gene in both GFP+ and GFP- cells
    gfp_pos_expr_gene = gfp_pos_expr[:, genes_of_interest.index(gene)]
    gfp_neg_expr_gene = gfp_neg_expr[:, genes_of_interest.index(gene)]
    
    # Combine the two groups (GFP+ and GFP-)
    all_expr = np.concatenate([gfp_pos_expr_gene, gfp_neg_expr_gene])
    all_labels = np.concatenate([np.ones(len(gfp_pos_expr_gene)), np.zeros(len(gfp_neg_expr_gene))])  # 1 = GFP+, 0 = GFP-

    # Calculate Pearson correlation
    corr, _ = pearsonr(all_expr, all_labels)
    correlations[gene] = corr

# Print the correlation results
for gene, corr in correlations.items():
    print(f"Correlation of {gene} expression with GFP status: {corr}")

# Optional: Plot expression of selected genes across GFP+ and GFP- cells
fig, axes = plt.subplots(1, len(genes_of_interest), figsize=(12, 5))

for i, gene in enumerate(genes_of_interest):
    if gene in adata.var_names:  # Ensure the gene is available in the main expression matrix
        # Get the expression values from the main expression matrix for the given gene
        expr_data = adata[:, gene].X.toarray().flatten()

        # Add expression data to a new column in obs for plotting
        adata.obs[gene] = expr_data

        # Create a boxplot using the expression data
        sns.boxplot(x='Sample', y=gene, data=adata.obs[adata.obs['Sample'].isin(['ASham_GFP', 'ASham_noGFP'])], ax=axes[i])
        axes[i].set_title(gene)
    else:
        axes[i].set_title(f"{gene} not found")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up data for correlation plot
genes_of_interest = ['C1qa', 'C1qb', 'C1qc']
gfp_labels = np.concatenate([
    np.ones(gfp_positive.shape[0]),   # 1 for GFP+
    np.zeros(gfp_negative.shape[0])   # 0 for GFP-
])

# Extract expression from adata.X (not adata.raw)
expr_data = {}
for gene in genes_of_interest:
    if gene in adata.var_names:
        expr_gfp_pos = adata[adata.obs['Sample'] == 'ASham_GFP', gene].X.toarray().flatten()
        expr_gfp_neg = adata[adata.obs['Sample'] == 'ASham_noGFP', gene].X.toarray().flatten()
        expr_data[gene] = np.concatenate([expr_gfp_pos, expr_gfp_neg])

# Create correlation plots
fig, axes = plt.subplots(1, len(genes_of_interest), figsize=(15, 5))

for i, gene in enumerate(genes_of_interest):
    sns.regplot(
        x=gfp_labels,
        y=expr_data[gene],
        ax=axes[i],
        scatter_kws={'s': 10, 'alpha': 0.6},
        line_kws={'color': 'red'}
    )
    corr, _ = pearsonr(expr_data[gene], gfp_labels)
    axes[i].set_title(f'{gene}\nPearson r = {corr:.2f}')
    axes[i].set_xlabel('GFP Status (1 = GFP+, 0 = GFP)')
    axes[i].set_ylabel('Expression')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Define genes of interest
genes_of_interest = ['C1qa', 'C1qb', 'C1qc', 'Gfp']  # Include Gfp

# Extract expression data (from raw if available)
expr_df = pd.DataFrame(
    adata.raw[:, genes_of_interest].X.toarray(),
    columns=genes_of_interest,
    index=adata.obs_names
)

# Drop any rows with missing values
expr_df.dropna(inplace=True)

# Calculate correlations with Gfp
correlations = {}
for gene in genes_of_interest:
    if gene != 'Gfp':
        corr, _ = pearsonr(expr_df[gene], expr_df['Gfp'])
        correlations[gene] = corr
        print(f"Correlation between {gene} and Gfp: {corr:.3f}")

# Plot gene expression vs Gfp
fig, axes = plt.subplots(1, len(correlations), figsize=(5 * len(correlations), 4))

if len(correlations) == 1:
    axes = [axes]

for i, gene in enumerate(correlations.keys()):
    sns.regplot(
        x='Gfp', y=gene, data=expr_df,
        ax=axes[i],
        scatter_kws={'s': 10, 'alpha': 0.6},
        line_kws={'color': 'red'}
    )
    axes[i].set_title(f'{gene} vs Gfp\nr = {correlations[gene]:.2f}')
    axes[i].set_xlabel('Gfp Expression')
    axes[i].set_ylabel(f'{gene} Expression')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Make sure genes of interest are defined
genes_of_interest = ['C1qa', 'C1qb', 'C1qc']

# Create a combined DataFrame for all cells
gfp_labels = adata.obs['Sample'].map({'ASham_GFP': 1, 'ASham_noGFP': 0}).astype(int)
correlation_data = pd.DataFrame({'GFP': gfp_labels})

# Extract gene expression from .raw
for gene in genes_of_interest:
    correlation_data[gene] = adata.raw[:, gene].X.toarray().flatten()

# Optional: check correlation values again
for gene in genes_of_interest:
    r, p = pearsonr(correlation_data[gene], correlation_data['GFP'])
    print(f"Correlation between {gene} and GFP: {r:.3f}, p-value: {p:.3e}")
```

```python
# Plot with regression line
fig, axes = plt.subplots(1, len(genes_of_interest), figsize=(5 * len(genes_of_interest), 4))

for i, gene in enumerate(genes_of_interest):
    ax = axes[i] if len(genes_of_interest) > 1 else axes
    sns.regplot(x='GFP', y=gene, data=correlation_data, ax=ax, logistic=False, scatter_kws={'s': 10})
    ax.set_title(f'{gene} vs GFP')
    ax.set_xlabel('GFP status (1=GFP+, 0=GFP)')
    ax.set_ylabel('Expression level')

plt.tight_layout()
plt.show()
```
