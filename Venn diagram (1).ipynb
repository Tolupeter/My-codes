```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import os

# Set working directory
os.chdir("P:/Tolulope/Manuscript/Yuan Analysis")

# Load DEG data from CSV files
df_ysham = pd.read_csv('YOV-GFP_vs_YOV-noGFP_M2LIKE.csv')
df_asham = pd.read_csv('AOV-GFP_vs_AOV-noGFP_M2LIKEI.csv')

# Ensure 'Symbol' column is in uppercase
df_ysham['Symbol'] = df_ysham['Symbol'].str.upper()
df_asham['Symbol'] = df_asham['Symbol'].str.upper()

# Define the conditions for upregulated and downregulated genes in each dataset
up_genes_ysham = set(df_ysham[(df_ysham['padj'] < 0.05) & (df_ysham['log2FoldChange'] > 0.5)]['Symbol'])
down_genes_ysham = set(df_ysham[(df_ysham['padj'] < 0.05) & (df_ysham['log2FoldChange'] < -0.5)]['Symbol'])

up_genes_asham = set(df_asham[(df_asham['padj'] < 0.05) & (df_asham['log2FoldChange'] > 0.5)]['Symbol'])
down_genes_asham = set(df_asham[(df_asham['padj'] < 0.05) & (df_asham['log2FoldChange'] < -0.5)]['Symbol'])

# Create Venn diagrams for upregulated and downregulated genes
plt.figure(figsize=(12, 6))

# Upregulated Genes
plt.subplot(1, 2, 1)
venn2([up_genes_ysham, up_genes_asham], ('YSham Upregulated', 'ASham Upregulated'))
plt.title('Upregulated Genes Comparison')

# Downregulated Genes
plt.subplot(1, 2, 2)
venn2([down_genes_ysham, down_genes_asham], ('YSham Downregulated', 'ASham Downregulated'))
plt.title('Downregulated Genes Comparison')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import os

# Set working directory
os.chdir("P:/Tolulope/Manuscript/Yuan Analysis")

# Load DEG data
df_ysham = pd.read_csv('YOV-GFP_vs_YOV-noGFP_M2LIKE.csv')
df_asham = pd.read_csv('AOV-GFP_vs_AOV-noGFP_M2LIKEI.csv')

# Normalize gene names
df_ysham['Symbol'] = df_ysham['Symbol'].str.upper()
df_asham['Symbol'] = df_asham['Symbol'].str.upper()

# DEG filters
up_genes_yov = set(df_ysham[(df_ysham['padj'] < 0.05) & (df_ysham['log2FoldChange'] > 0.5)]['Symbol'])
down_genes_yov = set(df_ysham[(df_ysham['padj'] < 0.05) & (df_ysham['log2FoldChange'] < -0.5)]['Symbol'])

up_genes_aov = set(df_asham[(df_asham['padj'] < 0.05) & (df_asham['log2FoldChange'] > 0.5)]['Symbol'])
down_genes_aov = set(df_asham[(df_asham['padj'] < 0.05) & (df_asham['log2FoldChange'] < -0.5)]['Symbol'])

# Plot
plt.figure(figsize=(12, 6))

# Upregulated genes
plt.subplot(1, 2, 1)
v1 = venn2([up_genes_yov, up_genes_aov], ('YOV', 'AOV'))
for patch in v1.patches:
    if patch: patch.set_alpha(0.5); patch.set_edgecolor('black'); patch.set_linewidth(1)
for text in v1.set_labels:
    text.set_fontsize(12)
for text in v1.subset_labels:
    if text: text.set_fontsize(10)
plt.title('Upregulated Genes in YOV vs AOV', fontsize=14)

# Downregulated genes
plt.subplot(1, 2, 2)
v2 = venn2([down_genes_yov, down_genes_aov], ('YOV', 'AOV'))
for patch in v2.patches:
    if patch: patch.set_alpha(0.5); patch.set_edgecolor('black'); patch.set_linewidth(1)
for text in v2.set_labels:
    text.set_fontsize(12)
for text in v2.subset_labels:
    if text: text.set_fontsize(10)
plt.title('Downregulated Genes in YOV vs AOV', fontsize=14)

plt.tight_layout()
plt.savefig("venn_comparison_YOV_vs_AOV.png", dpi=300, bbox_inches='tight')
plt.show()
```
